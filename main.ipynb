```python
from models.agent import run_automl_agent

def ask(question: str, thread_id: str = "notebook") -> None:
    print("Question:", question)
    result = run_automl_agent(question, thread_id=thread_id)
    print("\nAssistant response:")
    for msg in result["messages"]:
        content = getattr(msg, "content", None)
        if content:
            print(content)
```
